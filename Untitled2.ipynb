```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

# To d liu
means = [[2, 2], [9, 2], [4, 9]]
cov = [[2, 0], [0, 2]]
n_samples = 500
n_cluster = 3

X0 = np.random.multivariate_normal(means[0], cov, n_samples)
X1 = np.random.multivariate_normal(means[1], cov, n_samples)
X2 = np.random.multivariate_normal(means[2], cov, n_samples)
X = np.concatenate((X0, X1, X2), axis=0)

# V d liu ban u
plt.xlabel('x')
plt.ylabel('y')
plt.plot(X[:, 0], X[:, 1], 'bo', markersize=5)
plt.show()

# Khi to tm cm
def kmeans_init_centers(X, n_cluster):
    return X[np.random.choice(X.shape[0], n_cluster, replace=False)]

# Gn nhn
def kmeans_predict_labels(X, centers):
    D = cdist(X, centers)
    return np.argmin(D, axis=1)

# Cp nht tm
def kmeans_update_centers(X, labels, n_cluster):
    centers = np.zeros((n_cluster, X.shape[1]))
    for k in range(n_cluster):
        Xk = X[labels == k, :]
        centers[k, :] = np.mean(Xk, axis=0)
    return centers

# Kim tra hi t
def kmeans_has_converged(centers, new_centers):
    return (set(tuple(a) for a in centers) ==
            set(tuple(a) for a in new_centers))

# Hm v
def kmeans_visualize(X, centers, labels, n_cluster, title):
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)

    plt_colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']

    for i in range(n_cluster):
        data = X[labels == i]
        plt.plot(data[:, 0], data[:, 1], plt_colors[i] + '^', markersize=4,
                 label='cluster_' + str(i))
        plt.plot(centers[i][0], centers[i][1], plt_colors[i + 4] + 'o',
                 markersize=10, label='center_' + str(i))

    plt.legend()
    plt.show()

# Thut ton K-means
def kmeans(init_centers, init_labels, X, n_cluster):
    centers = init_centers
    labels = init_labels
    times = 0

    while True:
        labels = kmeans_predict_labels(X, centers)

        kmeans_visualize(
            X, centers, labels, n_cluster,
            'Assigned label for data at time = ' + str(times + 1)
        )

        new_centers = kmeans_update_centers(X, labels, n_cluster)

        if kmeans_has_converged(centers, new_centers):
            break

        centers = new_centers

        kmeans_visualize(
            X, centers, labels, n_cluster,
            'Update center position at time = ' + str(times + 1)
        )

        times += 1

    return (centers, labels, times)

# Chy chng trnh
init_centers = kmeans_init_centers(X, n_cluster)
print("Initial centers:\n", init_centers)

init_labels = np.zeros(X.shape[0])

kmeans_visualize(
    X, init_centers, init_labels, n_cluster,
    'Init centers in the first run. Assigned all data as cluster 0'
)

centers, labels, times = kmeans(init_centers, init_labels, X, n_cluster)
print('Done! Kmeans')
print('Times:', times)
print('Final centers:\n', centers)
```

```python
import numpy as np
from collections import Counter

# 1. Hm tnh khong cch Euclid
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

# 2. KNN vi trng s u (uniform)
def knn_uniform_predict(X_train, y_train, X_test, k):
    y_pred = []

    for x_test in X_test:
        # Tnh khong cch t x_test ti ton b X_train
        distances = [euclidean_distance(x_test, x_train) for x_train in X_train]

        # Ly index ca k im gn nht
        k_indices = np.argsort(distances)[:k]

        # Ly nhn ca k im 
        k_labels = y_train[k_indices]

        # B phiu theo a s
        most_common = Counter(k_labels).most_common(1)
        y_pred.append(most_common[0][0])

    return np.array(y_pred)

# 3. KNN vi trng s theo khong cch (distance weights)
def knn_distance_predict(X_train, y_train, X_test, k):
    y_pred = []
    eps = 1e-5  # trnh chia cho 0

    for x_test in X_test:
        distances = np.array([euclidean_distance(x_test, x_train) for x_train in X_train])

        # Ly k im gn nht
        k_indices = np.argsort(distances)[:k]
        k_labels = y_train[k_indices]
        k_distances = distances[k_indices]

        # Tnh trng s = 1 / distance
        weights = 1 / (k_distances + eps)

        # Cng trng s cho tng lp
        class_scores = {}
        for label, w in zip(k_labels, weights):
            if label in class_scores:
                class_scores[label] += w
            else:
                class_scores[label] = w

        # Chn lp c tng trng s ln nht
        pred_label = max(class_scores, key=class_scores.get)
        y_pred.append(pred_label)

    return np.array(y_pred)

    from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# To d liu
X, y = make_blobs(n_samples=200, centers=4, random_state=4)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

k = 5

# D on
y_pred_uniform = knn_uniform_predict(X_train, y_train, X_test, k)
y_pred_distance = knn_distance_predict(X_train, y_train, X_test, k)

# nh gi
print("Accuracy (uniform):", accuracy_score(y_test, y_pred_uniform))
print("Accuracy (distance):", accuracy_score(y_test, y_pred_distance))


```
