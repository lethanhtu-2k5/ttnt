```python
# -*- coding: utf-8 -*-
# -------------------- Đọc ma trận kề từ file --------------------
def read_adj_matrix(filename):
    matrix = []
    with open(filename, "r") as f:
        for line in f:
            if line.strip():
                matrix.append(list(map(int, line.strip().split())))
    return matrix

# -------------------- MAIN --------------------
filename = "graph.txt"  # File chứa ma trận kề 6x6
G = read_adj_matrix(filename)

# Tên các đỉnh của đồ thị.
node = "ABCDEF"
t_ = {}
for i in range(len(G)):
    t_[node[i]] = i

# Bậc của các đỉnh
degree = []
for i in range(len(G)):
    degree.append(sum(G[i]))

# Màu có thể sử dụng để tô cho các đỉnh
colorDict = {}
for i in range(len(G)):
    colorDict[node[i]] = ["Blue", "Red", "Yellow", "Green"]

# Sắp xếp các đỉnh theo thứ tự bậc giảm dần
sortedNode = []
indeks = []

# selection sort theo bậc
for i in range(len(degree)):
    _max = -1
    idx = -1
    for j in range(len(degree)):
        if j not in indeks:
            if degree[j] > _max:
                _max = degree[j]
                idx = j
    indeks.append(idx)
    sortedNode.append(node[idx])

# Phần xử lý tô màu (sử dụng màu ít nhất có thể)
theSolution = {}
for n in sortedNode:
    setTheColor = colorDict[n]
    theSolution[n] = setTheColor[0]
    adjacentNode = G[t_[n]]

    for j in range(len(adjacentNode)):
        if adjacentNode[j] == 1 and (setTheColor[0] in colorDict[node[j]]):
            colorDict[node[j]].remove(setTheColor[0])

# In kết quả từng đỉnh và màu đã tô tương ứng
print("Kết quả tô màu:")
for t, w in sorted(theSolution.items()):
    print("Đỉnh", t, "=", w)

# Ghi kết quả vào file
with open("result.txt", "w") as f:
    for t, w in sorted(theSolution.items()):
        f.write(f"Đỉnh {t} = {w}\n")

print("Đã ghi kết quả vào file coloring_result.txt")
```

```python
# -*- coding: utf-8 -*-

# -------------------- HÀM ĐỌC FILE --------------------
def read_adj_matrix(filename):
    matrix = []
    with open(filename, "r") as f:
        for line in f:
            if line.strip():
                matrix.append(list(map(int, line.strip().split())))
    return matrix

# -------------------- HÀM TÍNH BẬC --------------------
def compute_degree(G):
    degree = []
    for i in range(len(G)):
        degree.append(sum(G[i]))
    return degree

# -------------------- HÀM SẮP XẾP ĐỈNH --------------------
def sort_nodes_by_degree(node, degree):
    sortedNode = []
    indeks = []

    for i in range(len(degree)):
        _max = -1
        idx = -1
        for j in range(len(degree)):
            if j not in indeks:
                if degree[j] > _max:
                    _max = degree[j]
                    idx = j
        indeks.append(idx)
        sortedNode.append(node[idx])

    return sortedNode

# -------------------- HÀM TÔ MÀU --------------------
def color_graph(G, node, sortedNode):
    t_ = {}
    for i in range(len(G)):
        t_[node[i]] = i

    colorDict = {}
    for i in range(len(G)):
        colorDict[node[i]] = ["Blue", "Red", "Yellow", "Green"]

    theSolution = {}
    for n in sortedNode:
        setTheColor = colorDict[n]
        theSolution[n] = setTheColor[0]
        adjacentNode = G[t_[n]]

        for j in range(len(adjacentNode)):
            if adjacentNode[j] == 1 and (setTheColor[0] in colorDict[node[j]]):
                colorDict[node[j]].remove(setTheColor[0])

    return theSolution

# -------------------- HÀM GHI FILE --------------------
def save_result(solution, filename):
    with open(filename, "w") as f:
        for t, w in sorted(solution.items()):
            f.write(f"Đỉnh {t} = {w}\n")

# -------------------- MAIN --------------------
def main():
    filename = "graph.txt"
    G = read_adj_matrix(filename)

    node = "ABCDEF"
    degree = compute_degree(G)
    sortedNode = sort_nodes_by_degree(node, degree)
    solution = color_graph(G, node, sortedNode)

    # In kết quả
    print("Kết quả tô màu:")
    for t, w in sorted(solution.items()):
        print("Đỉnh", t, "=", w)

    # Ghi file
    save_result(solution, "result.txt")
    print("Đã ghi kết quả vào file result.txt")

# Chạy chương trình
main()
```

```python
from itertools import permutations

# -------------------- Ma trận chi phí 6x6 (khác nhau) --------------------
G = [
    [0, 12, 29, 22, 17, 19],
    [12, 0, 35, 27, 33, 24],
    [29, 35, 0, 31, 28, 26],
    [22, 27, 31, 0, 20, 23],
    [17, 33, 28, 20, 0, 21],
    [19, 24, 26, 23, 21, 0]
]

n = len(G)
cities = list(range(n))  # [0,1,2,3,4,5]

# -------------------- Brute Force TSP --------------------
min_cost = float('inf')
best_path = []

# Giữ thành phố 0 làm điểm xuất phát
for perm in permutations(cities[1:]):
    path = [0] + list(perm) + [0]  # Chu trình đi từ 0 → ... → 0
    cost = 0
    for i in range(len(path)-1):
        cost += G[path[i]][path[i+1]]
    if cost < min_cost:
        min_cost = cost
        best_path = path

# -------------------- Hiển thị kết quả --------------------
print("Chu trình tối ưu:", ' -> '.join(map(str, best_path)))
print("Chi phí tối thiểu:", min_cost)
```
